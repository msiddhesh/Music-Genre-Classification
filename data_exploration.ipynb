```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with the data
corr_matrix = features_30.drop(columns=["filename","label"]).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import numpy as np
```

```python
mask = (corr_matrix < 0.8) & (corr_matrix > -0.8)
filtered_corr_matrix = corr_matrix.mask(mask)
mask = np.eye(filtered_corr_matrix.shape[0], dtype=bool)
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1,mask=mask)
plt.title('Filtered Correlation Matrix (|correlation| > 0.5)')
plt.show()
```
