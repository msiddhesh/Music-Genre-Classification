```python
## import statements

import pandas as pd
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa as lb
import scipy as sp
import time 
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import PredefinedSplit
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
audio_feature_df = pd.read_csv('audio_feature_df.csv')
genre_names= ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']

## Train-Validation-Split

df_extracted_y = audio_feature_df["Label"]
df_extracted_x = audio_feature_df.drop("Label",axis = 1)

## Train 80%
## Test 10%
## Validation 8%

X_train_val, X_test, y_train_val, y_test = train_test_split(df_extracted_x,df_extracted_y,test_size = 0.20, stratify = df_extracted_y,random_state = 21)
X_train, X_val, y_train, y_val = train_test_split(X_train_val,y_train_val,test_size = 0.2,random_state = 21)
## Label Encoder
label_encoder = LabelEncoder().fit(y_train_val)

## We scale the data first
sc = StandardScaler().fit(X_train)

scaled_X_train = sc.transform( X_train )
scaled_X_val = sc.transform( X_val )
scaled_X_test = sc.transform( X_test )
scaled_X_train_val = sc.transform( X_train_val )
```

```python
sdjfan = [[1,2],[1,3],[1,1]]
```

```python
StandardScaler().fit_transform(sdjfan)
```

```python
### WITH PCA

pca = PCA().fit(scaled_X_train)
pca_X_train = pca.transform(scaled_X_train)

## Getting explained_variance_ratio
exp_var_pca = pca.explained_variance_ratio_
cummulative_var = np.cumsum(exp_var_pca)


plt.figure( figsize=(8,5) )
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cummulative_var)), cummulative_var, where='mid',label='Cumulative explained variance',color='r')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()


pca = PCA(n_components=0.95).fit(scaled_X_train)
pca_X_train = pca.transform(scaled_X_train)

print(f'Number of components {pca_X_train.shape[1]}')

pca_X_val = pca.transform( scaled_X_val )
pca_X_test = pca.transform( scaled_X_test )
pca_X_train_val = pca.transform( scaled_X_train_val )
```

```python
cummulative_var[:75]
```

```python
## Creating a predefined split so that we can feed it into the Grid Search

split_index = [-1]*len(pca_X_train) + [0]*len(pca_X_val)
X_grid = np.concatenate((pca_X_train, pca_X_val), axis=0)
y_grid = np.concatenate((y_train, y_val), axis=0)
pds = PredefinedSplit(test_fold = split_index)
```

```python
pds
```

#### Logistic Regression - OVR (one vs rest)

```python
## Logistic Regression - OVR (one vs rest)
## If we have 10 classes, we build a classifier for each of the class where the output is either it belongs to the class
## or it doesn't

param_grid = [
    { 'penalty': ['l2'],'C': [0.05,0.01,0.1,4,1,5,10,0.2],'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'] },
    { 'penalty': ['l1'],'C': [0.1,4,1,5,10,0.2],'solver':['liblinear','saga'] },
    { 'penalty': ['elasticnet'],'C': [0.1,4,1,5,10,0.2], 'l1_ratio':[0.25,0.5,0.75], 'solver':['saga'] },
]

logistic_regression_model = LogisticRegression(random_state=42, multi_class='ovr', n_jobs=-1,max_iter=1000)

gs = GridSearchCV(logistic_regression_model, cv = pds,param_grid=param_grid, scoring='accuracy')
gs.fit(X_grid,y_grid)
print( f' Best Hyperparameter: {gs.best_params_}' )
print( f' Best Score: {gs.best_score_}' )

final_logistic_regression_ovr = gs.best_estimator_.fit(pca_X_train_val,y_train_val)
y_test_pred_log_ovr =  final_logistic_regression_ovr.predict(pca_X_test)
print( f'Model Accuracy : { metrics.accuracy_score(y_test, y_test_pred_log_ovr) }' )


plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust font size for label

# Create a colorful heatmap
sns.heatmap(metrics.confusion_matrix(y_test,y_test_pred_log_ovr ), annot=True, fmt="d", cmap="inferno", cbar=False,
            xticklabels=genre_names, yticklabels=genre_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
## Logistic Regression : Multinomial 
## It means it apply softmax at the end 

param_grid = [
    { 'penalty': ['l2'],'C': [0.1,4,1,5,10,0.2],'solver':['lbfgs','newton-cg','sag','saga'] },
    { 'penalty': ['l1'],'C': [0.1,4,1,5,10,0.2],'solver':['saga'] },
    { 'penalty': ['elasticnet'],'C': [0.1,4,1,5,10,0.2], 'l1_ratio':[0.25,0.5,0.75], 'solver':['saga'] },
]

logistic_regression_model = LogisticRegression(random_state=42, multi_class='multinomial', n_jobs=-1, max_iter=1000)


gs = GridSearchCV(logistic_regression_model, cv = pds,param_grid=param_grid, scoring='accuracy')
gs.fit(X_grid,y_grid)
print( f' Best Hyperparameter: {gs.best_params_}' )
print( f' Best Score: {gs.best_score_}' )

final_logistic_regression_multinomial = gs.best_estimator_.fit(pca_X_train_val,y_train_val)
y_test_pred_log_multinomial =  final_logistic_regression_multinomial.predict(pca_X_test)
print( f'Model Accuracy : { metrics.accuracy_score(y_test, y_test_pred_log_multinomial) }' )


plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust font size for label

# Create a colorful heatmap
sns.heatmap(metrics.confusion_matrix(y_test,y_test_pred_log_multinomial ), annot=True, fmt="d", cmap="inferno", cbar=False,
            xticklabels=genre_names, yticklabels=genre_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

```python
## Support Vector Machine -ovr

param_grid = {
    'C' : [10,15,20],
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree' : [2,3,4],
    'gamma' : [0.0005,0.0001,0.001,0.01,0.1,1,2,5,10],
}
svc = SVC(random_state=42, decision_function_shape='ovr',break_ties=True, shrinking = False, class_weight='balanced')


gs = GridSearchCV(svc, cv = pds, param_grid=param_grid, scoring='accuracy')
gs.fit(X_grid,y_grid)
print( f' Best Hyperparameter: {gs.best_params_}' )
print( f' Best Score: {gs.best_score_}' )

final_svc_ovr = gs.best_estimator_.fit(pca_X_train_val,y_train_val)

y_test_pred_svc_ovr =  final_svc_ovr.predict(pca_X_test)
print( f'Model Accuracy : { metrics.accuracy_score(y_test, y_test_pred_svc_ovr) }' )


plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust font size for label

# Create a colorful heatmap
sns.heatmap(metrics.confusion_matrix(y_test,y_test_pred_svc_ovr ), annot=True, fmt="d", cmap="inferno", cbar=False,
            xticklabels=genre_names, yticklabels=genre_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt

# Define the hyperparameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300, 500],  # Number of trees in the forest
    'max_depth': [10, 20, 30, None],       # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],       # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],         # Minimum number of samples required at each leaf node
    'bootstrap': [True, False]             # Method for sampling data
}

# Initialize the RandomForestClassifier
random_forest_model = RandomForestClassifier(random_state=42)

# Use GridSearchCV to search for the best hyperparameters
gs_rf = GridSearchCV(random_forest_model, param_grid=param_grid_rf, cv=pds, scoring='accuracy')
gs_rf.fit(X_grid, y_grid)

# Display the best hyperparameters and the corresponding score
print(f'Best Hyperparameter: {gs_rf.best_params_}')
print(f'Best Score: {gs_rf.best_score_}')

# Train the final RandomForest model using the best hyperparameters
final_random_forest_model = gs_rf.best_estimator_.fit(pca_X_train_val, y_train_val)

# Make predictions on the test set
y_test_pred_rf = final_random_forest_model.predict(pca_X_test)

# Calculate and display the accuracy of the Random Forest model
print(f'Model Accuracy: {metrics.accuracy_score(y_test, y_test_pred_rf)}')

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)

# Create a colorful heatmap for the confusion matrix
sns.heatmap(metrics.confusion_matrix(y_test, y_test_pred_rf), annot=True, fmt="d", cmap="inferno", cbar=False,
            xticklabels=genre_names, yticklabels=genre_names)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
