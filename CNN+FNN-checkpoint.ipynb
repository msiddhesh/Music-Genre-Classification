```python
def prepare_image_numpy_array():

    all_file_paths = glob.glob('/content/drive/MyDrive/genre Classification/Spectograms/*/*')
    images_array = []

    for file in all_file_paths:

        img = img_to_array( load_img( file,target_size=(120,180) ) )
        images_array.append( img )

    return np.array( images_array )

```

```python
images = prepare_image_numpy_array()
features = pd.read_csv('/content/drive/MyDrive/genre Classification/audio_feature_df.csv')
y = features['Label']
features = features.drop(['Label'],axis = 1)
```

```python
train_feature_X, val_feature_X, train_Images_X, val_Images_X, y_train,y_val \
                            = train_test_split(train_feature_X, train_Images_X ,y_train, test_size=0.12, random_state=42)
```

```python
label_encoder = LabelEncoder().fit(y)
ohe_label_y_train = to_categorical( label_encoder.transform( y_train ) )
ohe_label_y_val = to_categorical( label_encoder.transform( y_val ) )
```

```python
input_features = keras.layers.Input( shape=train_feature_X.shape[1:])
x = keras.layers.Dense(units=32,activation=None,kernel_regularizer=l2())(input_features)
# x = keras.layers.Dropout(0.5)(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = keras.layers.Dense(units=32,activation=None,kernel_regularizer=l2())(x)
# x = keras.layers.Dropout(0.5)(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = keras.layers.Dense(units=32,activation=None,kernel_regularizer=l2())(x)
x = BatchNormalization()(x)
feature_dense_features = keras.layers.Activation('relu')(x)


input_images = keras.layers.Input( shape=train_Images_X[0].shape )
x = keras.layers.Conv2D(filters=32,kernel_size=3,activation=None,kernel_regularizer=l2())(input_images)
# x = keras.layers.Dropout(0.5)(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = keras.layers.MaxPool2D(pool_size=2)(x)
x = keras.layers.Conv2D(filters=64,kernel_size=5,activation=None,kernel_regularizer=l2())(x)
# x = keras.layers.Dropout(0.5)(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = keras.layers.MaxPool2D(pool_size=2)(x)
x = keras.layers.Conv2D(filters=128,kernel_size=5,activation=None,kernel_regularizer=l2())(x)
# x = keras.layers.Dropout(0.5)(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
x = keras.layers.MaxPool2D(pool_size=2)(x)
x = keras.layers.Flatten()(x)
x = keras.layers.Dense(units=128,activation=None,kernel_regularizer=l2(0.25))(x)
x = BatchNormalization()(x)
x = keras.layers.Activation('relu')(x)
image_dense_features = keras.layers.Dense(units=32,activation=None,kernel_regularizer=l2())(x)

concat_layer = keras.layers.concatenate([feature_dense_features,image_dense_features])
cfnn_dense_1 = keras.layers.Dense(units=32,activation='relu',kernel_regularizer=l2())(concat_layer)
cfnn_dense_2 = keras.layers.Dense(units=16,activation='relu',kernel_regularizer=l2())(cfnn_dense_1)
output_layer = keras.layers.Dense(units=10,activation='softmax')(cfnn_dense_2)

cnn_fnn = keras.models.Model(inputs=[input_features, input_images], outputs=[output_layer])
cnn_fnn.compile( loss= 'categorical_crossentropy', optimizer=keras.optimizers.Adadelta(learning_rate=0.01),metrics= ['accuracy'])
```

```python
fnn_history = cnn_fnn.fit(
        x=(train_feature_X,train_Images_X),
        y=ohe_label_y_train,
        validation_data=[(val_feature_X,val_Images_X),ohe_label_y_val],
        shuffle=True,
        batch_size=8,
        epochs=200
    )
```

```python

plt.style.use('ggplot')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

# Left Subplot (axes[0])
axes[0].plot(fnn_history.history['loss'], label='loss')
axes[0].plot(fnn_history.history['val_loss'], label='val loss')
axes[0].set_title("Loss vs Val_Loss")
axes[0].set_xlabel("Epochs")
axes[0].set_ylabel("Loss")
axes[0].legend()

# Right Subplot (axes[1])
axes[1].plot(fnn_history.history['accuracy'], label='accuracy')
axes[1].plot(fnn_history.history['val_accuracy'], label='val accuracy')
axes[1].set_title("Accuracy vs Val_Accuracy")
axes[1].set_xlabel("Epochs")
axes[1].set_ylabel("Accuracy")
axes[1].legend()

plt.tight_layout()  # Ensures that the subplots do not overlap
plt.show()
```

```python
train_feature_X, test_feature_X, train_Images_X, test_Images_X, y_train,y_test \
                            = train_test_split(features, images,y, test_size=0.2, random_state=42)
ohe_label_y_train = to_categorical( label_encoder.transform( y_train ) )
fnn_history = cnn_fnn.fit(
        x=(train_feature_X,train_Images_X),
        y=ohe_label_y_train,
        shuffle=True,
        batch_size=8,
        epochs=75
)
```

```python
y_pred = label_encoder.inverse_transform(np.argmax(cnn_fnn.predict((test_feature_X,test_Images_X)),axis=1))
print( f'Model Accuracy : { metrics.accuracy_score(y_test, y_pred ) }' )
```

```python
# Create a colorful heatmap
genre_names= ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']
sns.heatmap(metrics.confusion_matrix(y_test,y_pred ), annot=True, fmt="d", cmap="inferno", cbar=False,
            xticklabels=genre_names, yticklabels=genre_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
