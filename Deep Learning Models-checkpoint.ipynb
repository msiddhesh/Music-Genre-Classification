```python
## import statements

import pandas as pd
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa as lb
import scipy as sp
import time 
import seaborn as sns
import copy


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam,SGD,RMSprop,Adagrad
from tensorflow.keras.metrics import categorical_accuracy
from keras.utils import to_categorical
from tensorflow.keras.regularizers import L1L2
import keras_tuner as kt


from sklearn import metrics
```

```python
audio_feature_df = pd.read_csv('audio_feature_df.csv')
genre_names= ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']

## Train-Validation-Split

y = audio_feature_df['Label']
X = audio_feature_df.drop( 'Label',axis=1 )

## Train 80%
## Test 12%
## Validation 8%
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.12, random_state=42, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.12, random_state=42, stratify=y_train_val)

## Label Encoder
label_encoder = LabelEncoder().fit(y_train_val)

## ohe: one hot encoded
ohe_label_y_train = to_categorical( label_encoder.transform( y_train ) )
ohe_label_y_train_val = to_categorical( label_encoder.transform( y_train_val ) )
ohe_label_y_val = to_categorical( label_encoder.transform( y_val ) )

## We scale the data first
sc = StandardScaler().fit(X_train)

scaled_X_train = sc.transform( X_train )
scaled_X_val = sc.transform( X_val )
scaled_X_test = sc.transform( X_test )
scaled_X_train_val = sc.transform( X_train_val )


```

```python
### WITH PCA

pca = PCA().fit(scaled_X_train)
pca_X_train = pca.transform(scaled_X_train)

## Getting explained_variance_ratio
exp_var_pca = pca.explained_variance_ratio_
cummulative_var = np.cumsum(exp_var_pca)


plt.figure( figsize=(8,5) )
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cummulative_var)), cummulative_var, where='mid',label='Cumulative explained variance',color='r')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()


pca = PCA(n_components=0.95).fit(scaled_X_train)
pca_X_train = pca.transform(scaled_X_train)

print(f'Number of components {pca_X_train.shape[1]}')

pca_X_val = pca.transform( scaled_X_val )
pca_X_test = pca.transform( scaled_X_test )
pca_X_train_val = pca.transform( scaled_X_train_val )
```

```python
### Fullconnected Feedforward Nueral Network
```

```python
## Building the model

def fnn():

    input_layer = keras.layers.Input( shape=pca_X_train.shape[1:] )
    hidden_layer_1 = keras.layers.Dense( 32, activation='relu' )(input_layer)
    hidden_layer_2 = keras.layers.Dense( 32, activation='relu' )(hidden_layer_1)
    hidden_layer_3 = keras.layers.Dense( 32, activation='relu' )(hidden_layer_2)
    output_layer = keras.layers.Dense( 10, activation='softmax')(hidden_layer_3)

    fnn_model =  keras.models.Model(inputs=[input_layer],outputs=[output_layer])

    fnn_model.compile( 
        loss= 'categorical_crossentropy', 
        optimizer= Adam(learning_rate=0.0001),
        metrics= ['accuracy']
    )

    fnn_history = fnn_model.fit(
        x=pca_X_train,
        y=ohe_label_y_train,
        validation_data= (pca_X_val,ohe_label_y_val),
        shuffle=True,
        batch_size=20,
        epochs=1000,
        callbacks=[tf.keras.callbacks.EarlyStopping('val_loss', patience=50)] ,
        verbose=0
    )

    n_epochs_best = np.argmax( fnn_history.history['val_loss'] )

    fnn_model.fit(
        x=pca_X_train_val,
        y=ohe_label_y_train_val,
        batch_size=20,
        epochs=n_epochs_best,
        verbose=0
    )

    y_test_pred_fnn = label_encoder.inverse_transform(np.argmax(fnn_model.predict( pca_X_test ),axis=1))
    return metrics.accuracy_score(y_test,y_test_pred_fnn)

accuracy = []
for i in range(50):
    accuracy.append( fnn() )
    print( f'{i+1} run complete')
```

```python
print( np.mean(accuracy) )
np.std(accuracy)
```

```python
def fnn_tuned():

    model_optimizer = Adam( 
                learning_rate=0.0005,
                beta_1 = 0.9,
                beta_2 = 0.8,
                epsilon = 1e-7,
            )

    fnn_tuned_model = Sequential()

    ## Input layer
    fnn_tuned_model.add( Dropout(rate=0.2,input_shape=pca_X_train.shape[1:],seed=42))

    ## Dense Layer 1
    # model.add(BatchNormalization())
    fnn_tuned_model.add(Dropout(rate=0.2,seed=42))
    fnn_tuned_model.add( Dense(128, activation='relu' ,kernel_regularizer=L1L2( l1 = 0.001,l2 = 0.001))) 

    ## Dense Layer 2
    # model.add(BatchNormalization())
    fnn_tuned_model.add( Dense(96, activation='sigmoid',kernel_regularizer=L1L2( l1 = 0.001,l2 = 0.01) ) )

    ## Dense Layer 3
    # model.add(BatchNormalization())
    fnn_tuned_model.add( Dense(32, activation='tanh',kernel_regularizer=L1L2( l1 = 0.001,l2 = 0.001) ) )

    ## Dense Layer 4
    # model.add(BatchNormalization())
    fnn_tuned_model.add(Dropout(rate=0.4,seed=42))
    fnn_tuned_model.add( Dense(128, activation='tanh',kernel_regularizer=L1L2( l1 = 0.001,l2 = 0.01) ) )

    ## Outputlayer
    fnn_tuned_model.add( Dense(10, activation='softmax') )

    fnn_tuned_model.compile( 
        optimizer=model_optimizer,
        loss= 'categorical_crossentropy', 
        metrics= ['accuracy']    
    )

    fnn_tuned_model_history = fnn_tuned_model.fit(
        x=pca_X_train,
        y=ohe_label_y_train,
        validation_data= (pca_X_val,ohe_label_y_val),
        shuffle=True,
        batch_size=20,
        epochs=1000,
        callbacks=[tf.keras.callbacks.EarlyStopping('val_loss', patience=50)] ,
        verbose=0
    )

    n_epochs_best = np.argmax( fnn_tuned_model_history.history['val_loss'] )

    fnn_tuned_model.fit(
        x=pca_X_train_val,
        y=ohe_label_y_train_val,
        shuffle=True,
        batch_size=20,
        epochs=n_epochs_best, 
        verbose=0
    )

    
accuracy_tuned = []
for i in range(50):
    accuracy_tuned.append( fnn_tuned() )
    print( f'{i+1} run complete')


```
